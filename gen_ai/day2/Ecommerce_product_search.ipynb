```python
# Scenario: E-Commerce Product Search
# - Problem: Customers search using phrases like "cheap running shoes"
#   while product descriptions use terms like "budget-friendly athletic footwear".
# - Solution: Use embeddings to capture semantic meaning so similar concepts
#   can be matched even when the wording is different.


from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual retrieval
# "sentence-transformers/e5-large-v2" → strong performance

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Product catalog (marketing descriptions)
products = [
    "Budget-friendly athletic footwear",
    "High-performance professional running shoes",
    "Comfortable casual walking sneakers",
    "Premium leather formal shoes"
]

# Customer search query (everyday language)
query = "cheap running shoes"

# Convert product descriptions and query into embeddings
product_embeddings = model.encode(products)
query_embedding = model.encode([query])

# Compute similarity between query and products
similarity_scores = cosine_similarity(query_embedding, product_embeddings)

# Find the most relevant product
best_match_index = similarity_scores.argmax()

print(f"Model: {model_name}")
print("Customer Query:", query)

print("\nMost Relevant Product:")
print(products[best_match_index])

print("\nSimilarity Scores:", similarity_scores)
```
