```python
#  Scenario: University Library Search
# Imagine you’re building a smart search system for a university library.
# - Problem: Students often type queries like “fast dog running” or “canine sprinting”, but
# the library catalog has books titled “The dog runs fast”. Traditional keyword search misses these connections.
# - Solution: Use embeddings to capture semantic meaning instead of exact words.


from sentence_transformers import SentenceTransformer

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual + dense/sparse retrieval
# "sentence-transformers/e5-large-v2" → 1024 dims, top performer on MTEB

model_name = "all-MiniLM-L6-v2"   # swap this to test other models
model = SentenceTransformer(model_name)

texts = [
    "The dog runs fast",
    "A quick canine sprints",
    "Python is a programming language"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")   # e.g. (3, 384) for MiniLM
print(f"First vector sample: {embeddings[0][:10]}")  # show first 10 values
```

```python
# Scenario: Hospital Patient Record Search
# - Problem: Doctors and nurses often type queries like “heart attack” or “chest pain emergency”, but the hospital’s electronic health record (EHR) system stores diagnoses as “acute myocardial infarction”. Traditional keyword search fails because medical staff use everyday language while records use clinical terminology.
# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that “heart attack” and “acute myocardial infarction” are related concepts. This allows the search system to return the right patient records even when the query wording doesn’t match the stored terminology.
# Scenario: Hospital Patient Record Search
# - Problem: Doctors and nurses type queries like "heart attack" or "chest pain emergency"
#   while hospital records store diagnoses using clinical terms like
#   "acute myocardial infarction".
# - Solution: Use embeddings so that semantically similar medical concepts
#   can be matched even if the wording is different.


from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual + dense/sparse retrieval
# "sentence-transformers/e5-large-v2" → 1024 dims, top performer on MTEB

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Hospital patient records (clinical terminology)
records = [
    "Patient diagnosed with acute myocardial infarction",
    "Patient suffering from hypertension",
    "Patient admitted with severe asthma attack",
    "Patient with diabetes mellitus type 2"
]

# Doctor's query using everyday language
query = "heart attack emergency"

# Convert records and query to embeddings
record_embeddings = model.encode(records)
query_embedding = model.encode([query])

# Compute similarity
similarity_scores = cosine_similarity(query_embedding, record_embeddings)

# Find the most relevant record
best_match_index = similarity_scores.argmax()

print(f"Model: {model_name}")
print("Doctor Query:", query)
print("\nMost Relevant Patient Record:")
print(records[best_match_index])
print("\nSimilarity Scores:", similarity_scores)



```

```python
# Scenario: E-Commerce Product Search
# - Problem: Customers search using phrases like "cheap running shoes"
#   while product descriptions use terms like "budget-friendly athletic footwear".
# - Solution: Use embeddings to capture semantic meaning so similar concepts
#   can be matched even when the wording is different.


from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual retrieval
# "sentence-transformers/e5-large-v2" → strong performance

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Product catalog (marketing descriptions)
products = [
    "Budget-friendly athletic footwear",
    "High-performance professional running shoes",
    "Comfortable casual walking sneakers",
    "Premium leather formal shoes"
]

# Customer search query (everyday language)
query = "cheap running shoes"

# Convert product descriptions and query into embeddings
product_embeddings = model.encode(products)
query_embedding = model.encode([query])

# Compute similarity between query and products
similarity_scores = cosine_similarity(query_embedding, product_embeddings)

# Find the most relevant product
best_match_index = similarity_scores.argmax()

print(f"Model: {model_name}")
print("Customer Query:", query)

print("\nMost Relevant Product:")
print(products[best_match_index])

print("\nSimilarity Scores:", similarity_scores)
```
