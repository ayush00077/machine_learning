```python
# Scenario: Hospital Patient Record Search
# - Problem: Doctors and nurses often type queries like вЂњheart attackвЂќ or вЂњchest pain emergencyвЂќ, but the hospitalвЂ™s electronic health record (EHR) system stores diagnoses as вЂњacute myocardial infarctionвЂќ. Traditional keyword search fails because medical staff use everyday language while records use clinical terminology.

# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that вЂњheart attackвЂќ and вЂњacute myocardial infarctionвЂќ are related concepts. This allows the search system to return the right patient records even when the query wording doesnвЂ™t match the stored terminology.
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Choose from popular models:
# "all-MiniLM-L6-v2" в†’ 384 dims, lightweight, fast
# "all-mpnet-base-v2" в†’ 768 dims, stronger semantic understanding
# "BAAI/bge-m3" в†’ 1024 dims, multilingual + dense/sparse retrieval
# "sentence-transformers/e5-large-v2" в†’ 1024 dims, top performer on MTEB

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Hospital patient records (clinical terminology)
records = [
    "Patient diagnosed with acute myocardial infarction",
    "Patient suffering from hypertension",
    "Patient admitted with severe asthma attack",
    "Patient with diabetes mellitus type 2"
]

# Doctor's query using everyday language
query = "heart attack emergency"

# Convert records and query to embeddings
record_embeddings = model.encode(records)
query_embedding = model.encode([query])

# Compute similarity
similarity_scores = cosine_similarity(query_embedding, record_embeddings)

# Find the most relevant record
best_match_index = similarity_scores.argmax()

print(f"Model: {model_name}")
print("Doctor Query:", query)
print("\nMost Relevant Patient Record:")
print(records[best_match_index])
print("\nSimilarity Scores:", similarity_scores)


```
